## 2. One hot encoding with many categories in a feature

```python
df=pd.read_csv('mercedes.csv')
```

```python
df.head()
```

```python
df=pd.read_csv('mercedes.csv',usecols=['X0','X1','X2','X3','X4','X5','X6'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
for i in df.columns:
    print(df[i].value_counts())
```

```python
for i in df.columns:
    print(len(df[i].unique()))
```

```python
df.X1.value_counts(ascending=False).head(10)
```

```python
list_10=df.X1.value_counts(ascending=False).head(10).index
list_10
```

```python
list_10=list(list_10)
```

```python
list_10
```

```python
for categories in list_10:
    df[categories]=np.where(df['X1']==categories,1,0)
```

```python
list_10.append('X1')
```

```python
df[list_10]
```

```python
## Performing one hot encoding on X0 feature in the df dataset
```

```python
df['X0'].value_counts(ascending=False).head(10)
```

```python
X0_list_10=df['X0'].value_counts(ascending=False).head(10).index
X0_list_10=list(X0_list_10)
```

```python
X0_list_10
```

```python
pd.get_dummies(X0_list_10)
```
